```python
import pandas as pd
import numpy as np
nilai_kosong=['?','-',' ']
df = pd.read_csv('test.csv',na_values=nilai_kosong)
df.head()
```

```python
dftrain= pd.read_csv('train.csv',na_values=nilai_kosong)
dftrain.head()
```

```python
dftrain=pd.get_dummies(dftrain,columns=['Kelas Pekerja','Status Perkawinan','Pekerjaan','Jenis Kelamin'])
dftrain.head()
```

```python
obj_dict={
    '<=7jt':0,
    '>7jt':1
}
dftrain['Gaji']=dftrain['Gaji'].replace(obj_dict)
```

```python
obj_dict2={
    '1st-4th':0,
    '5th-6th':1,
    '7th-8th':2,
    '9th':3,
    '10th':4,
    '11th':5,
    '12th':6,
    'SD':7,
    'SMA':8,
    'Sekolah Professional':9,
    'D3':10,
    'D4':11,
    'Pendidikan Tinggi':12,
    'Sarjana':13,
    'Master':14,
    'Doktor':15
}
dftrain['Pendidikan']=dftrain['Pendidikan'].replace(obj_dict2)
```

```python
df=pd.get_dummies(df,columns=['Kelas Pekerja','Status Perkawinan','Pekerjaan','Jenis Kelamin'])
df.head()
```

```python
obj_dict2={
    '1st-4th':0,
    '5th-6th':1,
    '7th-8th':2,
    '9th':3,
    '10th':4,
    '11th':5,
    '12th':6,
    'SD':7,
    'SMA':8,
    'Sekolah Professional':9,
    'D3':10,
    'D4':11,
    'Pendidikan Tinggi':12,
    'Sarjana':13,
    'Master':14,
    'Doktor':15
}
df['Pendidikan']=df['Pendidikan'].replace(obj_dict2)
```

```python
X_test= df.drop ('id', axis=1)
X_test.head()
```

```python
X_train= dftrain.drop ('id', axis=1)
X_train.head()
```

```python
X_train2 = X_train.drop('Gaji', axis=1)
X_train2.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
y = dftrain['Gaji']
model_tree = DecisionTreeRegressor(min_weight_fraction_leaf=0.1)
model_tree.fit(X_train2, y)
```

```python
y_test=model_tree.predict(X_test)
```
